# Echoes of the Past – Conversing with Historical Voices Using Large Language Models

## 1. Idea Summary

"**Echoes of the Past**" is a creative application that uses a large language model (LLM) to simulate conversations between users and historical figures. It is designed to enhance engagement with history and cultural heritage by allowing users to ask questions and receive responses as if they were speaking directly to someone from a specific historical context.

Unlike traditional educational tools, this application includes both well-known and marginalised perspectives—such as colonised peoples, women, workers, and minority voices—supporting inclusive and decolonised approaches to learning.

**Target Audience**: Students, educators, and the general public interested in history, literature, and cultural studies within the arts and humanities.

---

## 2. Interaction Design

### Step-by-step Interaction:

1. **Select a Historical Context**  
   Example options:
   - The May Fourth Movement (China, 1919)
   - Caribbean Slave Rebellions (18th century)
   - The Civil Rights Movement (USA, 1960s)
   - The Partition of India (1947)

2. **Choose a Historical Figure or Voice**  
   Options may include:
   - Well-known figures (e.g., Martin Luther King Jr.)
   - Marginalised voices (e.g., enslaved person, domestic servant, textile worker)

3. **Initiate a Dialogue**  
   The user types questions or prompts. The LLM responds in a tone and manner consistent with the selected figure's historical background and worldview.

4. **Access Additional Features**  
   - Contextual background materials for each conversation  
   - Toggle between "mainstream" and "decolonised" perspectives  
   - Conversation summary for classroom use  
   - Optional voice input and multi-language support

### Interaction Diagram (Text-Based Example):

## 3. Value of the Application

### a) Educational Value (Learning)
- Promotes critical thinking and deeper historical understanding.
- Encourages users to explore different viewpoints through interaction.
- Supports personalised and reflective learning experiences.

### b) Engagement
- Makes history feel immediate and personal.
- Increases retention by encouraging active participation.
- Appeals to younger, digitally-native audiences.

### c) Inclusion
- Gives visibility to underrepresented narratives.
- Encourages users to challenge dominant historical perspectives.
- Supports decolonisation efforts in education.

---

## 4. Critical Reflection

### a) Potential Positive Impacts
- **Cultural Awareness**: Helps users understand the lived experiences of people from different times and backgrounds.  
- **Empathy Building**: Encourages users to emotionally connect with historical figures.  
- **Support for Inclusive Learning**: Provides access to alternative viewpoints often missing in traditional history education.

### b) Possible Risks
- **Historical Inaccuracy**: AI may generate incorrect or biased content.  
- **Oversimplification**: Dialogue format may reduce complex topics to short answers.  
- **False Credibility**: Users may mistake generated responses as factual or authoritative.  
- **Access Limitations**: Not all users may have equal access to required technologies or language options.

### c) Mitigation Strategies
- Display clear disclaimers with each AI-generated response.
- Provide verified background resources linked to conversations.
- Involve historians and cultural experts in system design.
- Ensure accessibility through screen reader compatibility and multilingual support.

---

## 5. Conclusion

"Echoes of the Past" is a powerful example of how large language models can be used creatively in the arts and humanities. It enables users to learn history through conversation, fosters empathy and curiosity, and challenges one-sided narratives by including diverse voices.

By combining thoughtful design with responsible use of AI, this application can offer a meaningful, inclusive learning experience. Its success depends on critical oversight, collaboration with scholars, and a strong focus on ethical engagement.
